# Maplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange (0,10,0.1) # Datos en x
y=2*np.sin(x) # Datos en y)
#plt.title('Señal en Frecunecia')   # Titulo
plt.title('Señal en Frecunecia',fontsize=12)

plt.xlabel('Tiempo[s]')          # Eje x 
plt.ylabel('Amplitud')           # Eje y
plt.plot(x,y,)  # Grafica
plt.grid()     #Rejilla
plt.legend(['2sin(t)'],loc=3)    # Leyenda loc=lugar
plt.show() # Elimino mensaje de instancia

```

```python
plt.rcParams.update({'font.size':14}) # Modifica tamaño de los ejes
```

Multiples Series 

```python
x= np.arange (0,10,0.1) # Datos en x
y=2*np.sin(x) # Datos en y)
z= np.cos(4*x) # Datos en Z
plt.title('Señal en Frecunecia',fontsize =16)   # Titulo
#plt.title('Señal en Frecunecia',fonsize=16)

plt.xlabel('Tiempo[s]')          # Eje x 
plt.ylabel('Amplitud')           # Eje y
#plt.plot(x,y)  # Grafica Seno
#plt.plot(x,z)  # Graficaa Coseno   
plt.plot(x,y,x,z) # AMBAS GGRAFICAS
plt.grid()     #Rejilla
plt.legend(['2sin(t)','cos(4t)'],loc=3)    # Leyenda loc=lugar
plt.show() # Elimino mensaje de instancia
```

```python
plt.plot(x,y) # Seno
plt.figure()  # Divide los Graficos
plt.plot(x,z) #Coseno
```

```python
plt.subplot(211)
plt.plot(x,y)
plt.subplot(212)
plt.plot(x,z)


#Primera Casilla = Filas
#Segunda Casilla = Columnas
#Tercera Casilla = El Axel 
```

Linea, Color, Estilo

```python
plt.plot(x,y,'-r',x,z,'--k')
plt.show()
```

```python
plt.plot(x,y,'4m',x,z,':c')
plt.show()
```

Tamaño de la Linea

```python
plt.plot(x,y,'-r',x,z,'--k',linewidth=10)
plt.show()
```

Escalonamineto de Ejes y Anotaciones

```python
fig,ax=plt.subplots()
plt.plot(x,y,'-r',x,z,'--k',linewidth=6)
ax.axis([3,7,-1,1])
plt.show()
```

Ubicar Texto en el Grafico

```python
fig,ax=plt.subplots()
plt.plot(x,y,'-r',x,z,'--k',linewidth=6)
ax.text(4,1.5,'Funciones en Frecuencia',fontsize=12)
plt.show()
```

Agregar Simbolos


```python

plt.plot(x,y,'-r',x,z,'--k',linewidth=6)
plt.title('Funcion y=Sin($\\alpha$ t)')
```

Grafico de Dispersion

```python
x=np.arange(0,100,5)
y=np.random.random(len(x))
plt.scatter(x,y,color='#ff0001',s=15)
plt.show()
```
